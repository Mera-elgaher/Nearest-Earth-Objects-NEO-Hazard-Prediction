# 2-Data preprossing

**Feature Engineering**

```python
df = pd.get_dummies(df, drop_first=True)
```

```python
X = df.drop(columns=['is_hazardous'])
y = df['is_hazardous']
```

```python
print("Class distribution before SMOTE:\n", y.value_counts())
```

**Handling Imbalanced Classes**

```python
X = df.drop(["neo_id", "name", "orbiting_body", "is_hazardous"], axis=1)

y = df["is_hazardous"]
```

```python
somte = SMOTE(random_state=42)

X, y = somte.fit_resample(X, y)

y.value_counts()
```

 **Train**-**Test** **Split**

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, random_state=42, stratify=y, test_size=0.2)
```

**Scaling**

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(x_train)
X_test = scaler.transform(x_test)
```

# Model 1 Training&Evaluation with RandomForest:

```python
modelRandomForest = RandomForestClassifier(random_state=42)
modelRandomForest.fit(x_train, y_train)
```

```python
y_pred = modelRandomForest.predict(x_test)
```

```python
print("Classification Report:\n", classification_report(y_test, y_pred))
print("ROC AUC Score:", roc_auc_score(y_test, y_pred))
```

```python
conf_matrix = confusion_matrix(y_test, y_pred)
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()
```

**ROC Curve for Random Forest**

```python

fpr, tpr, _ = roc_curve(y_test, modelRandomForest.predict_proba(x_test)[:,1])
roc_auc = auc(fpr, tpr)

plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```

# Model 2 Training & Evaluation with Gradient Boosting:

```python

gb_model = GradientBoostingClassifier(random_state=42)
gb_model.fit(X_train, y_train)
gb_y_pred = gb_model.predict(X_test)
```

```python

print("GradientBoosting Classification Report:\n", classification_report(y_test, gb_y_pred))
print("GradientBoosting ROC AUC Score:", roc_auc_score(y_test, gb_y_pred))
gb_conf_matrix = confusion_matrix(y_test, gb_y_pred)
sns.heatmap(gb_conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.title('GradientBoosting Confusion Matrix')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()


```

**ROC Curve for Gradient Boosting**

```python

gb_fpr, gb_tpr, _ = roc_curve(y_test, gb_model.predict_proba(X_test)[:,1])
gb_roc_auc = auc(gb_fpr, gb_tpr)
plt.figure()
plt.plot(gb_fpr, gb_tpr, color='blue', lw=2, label='GradientBoosting ROC curve (area = %0.2f)' % gb_roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```
